```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('.\Documents\Internship\insurance.csv') 
df.head()  	
  
```

```python

plt.figure(figsize=(10, 6))
plt.bar(df.Gender, df.bmi)  
plt.xlabel("Gender")
plt.ylabel("Bmi")
plt.title("Distribution of Categorical data Genders in a Population")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df.age)  
plt.xlabel("Age")
plt.title("Distribution of Continous data Ages in a Population")
plt.show()
```
